```python
import pandas as pd
```

```python
jan_df = pd.read_csv('raw data/JC-201901-citibike-tripdata.csv')
jan_df
```

```python
feb_df = pd.read_csv('raw data/JC-201902-citibike-tripdata.csv')
mar_df = pd.read_csv('raw data/JC-201903-citibike-tripdata.csv')
apr_df = pd.read_csv('raw data/JC-201904-citibike-tripdata.csv')
may_df = pd.read_csv('raw data/JC-201905-citibike-tripdata.csv')
jun_df = pd.read_csv('raw data/JC-201906-citibike-tripdata.csv')
jul_df = pd.read_csv('raw data/JC-201907-citibike-tripdata.csv')
aug_df = pd.read_csv('raw data/JC-201908-citibike-tripdata.csv')
sep_df = pd.read_csv('raw data/JC-201909-citibike-tripdata.csv')
oct_df = pd.read_csv('raw data/JC-201910-citibike-tripdata.csv')
nov_df = pd.read_csv('raw data/JC-201911-citibike-tripdata.csv')
dec_df = pd.read_csv('raw data/JC-201912-citibike-tripdata.csv')
```

```python
year2019 = [jan_df, feb_df, mar_df, apr_df, may_df, jun_df, jul_df, aug_df, sep_df, oct_df, nov_df, dec_df]
year2019
```
